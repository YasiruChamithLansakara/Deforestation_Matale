## Deforestation Analysis - Comprehensive Guide

This notebook performs detailed monthly deforestation analysis for Matale district, Sri Lanka using Landsat satellite data. 

### üéØ **Analysis Overview**

**Purpose**: Track forest loss and deforestation patterns over time using satellite imagery to understand:
- Monthly deforestation trends
- Seasonal variations in forest loss detection
- Long-term forest cover change patterns
- Deforestation change detection between time periods

**Data Source**: Landsat Surface Reflectance data
- **Temporal Coverage**: Multi-year analysis (adjust `YEARS` range as needed)
- **Spatial Resolution**: 30m pixels
- **Processing**: Monthly scenes processed individually
- **Area of Interest**: Matale district polygon

### üìä **Key Outputs**
1. **Monthly Statistics**: CSV file with deforested area measurements for each scene
2. **Temporal Visualizations**: Trend charts showing deforestation over time
3. **Raster Products**: NDVI, NDBI, and deforestation classification maps for each date
4. **Change Maps**: Forest loss/recovery between first and last scenes
5. **Seasonal Analysis**: Understanding seasonal variations in deforestation detection

```python
import os
import numpy as np
import rasterio
from rasterio.mask import mask
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
import glob
from datetime import datetime

# CONFIGURATION
BASE_DIR = r"D:\Satellite Image Processing\Satellite Images"
AOI_PATH = r"D:\Satellite Image Processing\Deforestation_Matale\AOI\matale_District_AOI.geojson"
OUTPUT_DIR = r"D:\Satellite Image Processing\Deforestation_Matale\Processed_Monthly"
os.makedirs(OUTPUT_DIR, exist_ok=True)

YEARS = range(2013, 2026)
BANDS = {
    "blue": "SR_B2.TIF",
    "green": "SR_B3.TIF", 
    "red": "SR_B4.TIF",
    "nir": "SR_B5.TIF",
    "swir": "SR_B6.TIF",
    "qa": "QA_PIXEL.TIF"  # Cloud/shadow QA band
}

# LOAD AOI
aoi = gpd.read_file(AOI_PATH)
print(f"‚úì AOI loaded: {len(aoi)} polygon(s) in {aoi.crs}")

# ============================================================
# CLOUD & SHADOW MASKING USING QA_PIXEL BAND
# ============================================================

def decode_qa_pixel(qa_band):
    """
    Decode Landsat QA_PIXEL band to create cloud/shadow mask.
    Returns True for valid pixels, False for cloudy/shadow pixels.
    
    QA_PIXEL bit flags (Landsat Collection 2):
    - Bit 0: Fill
    - Bit 1: Dilated Cloud
    - Bit 2: Cirrus (high confidence)
    - Bit 3: Cloud
    - Bit 4: Cloud Shadow
    - Bit 5: Snow
    - Bit 6: Clear (0=No, 1=Yes)
    - Bit 7: Water
    """
    # Create mask for clear pixels
    fill_mask = (qa_band & (1 << 0)) == 0  # Not fill
    cloud_mask = (qa_band & (1 << 3)) == 0  # Not cloud
    cloud_shadow_mask = (qa_band & (1 << 4)) == 0  # Not shadow
    dilated_cloud_mask = (qa_band & (1 << 1)) == 0  # Not dilated cloud
    cirrus_mask = (qa_band & (1 << 2)) == 0  # Not cirrus
    
    # Combine all masks - pixel is valid only if all conditions are met
    valid_mask = fill_mask & cloud_mask & cloud_shadow_mask & dilated_cloud_mask & cirrus_mask
    
    return valid_mask

def clip_to_aoi_with_qa(raster_path, qa_path, aoi_geom, target_crs):
    """
    Clip raster to AOI and apply cloud/shadow mask from QA band.
    Returns masked array with invalid pixels set to NaN.
    """
    with rasterio.open(raster_path) as src:
        aoi_reprojected = aoi_geom.to_crs(target_crs)
        out_img, out_transform = mask(src, aoi_reprojected.geometry, crop=True)
        out_meta = src.meta.copy()
        out_meta.update({
            "driver": "GTiff",
            "height": out_img.shape[1],
            "width": out_img.shape[2],
            "transform": out_transform
        })
    
    # Apply cloud mask if QA band is available
    if qa_path and os.path.exists(qa_path):
        with rasterio.open(qa_path) as qa_src:
            qa_img, _ = mask(qa_src, aoi_reprojected.geometry, crop=True)
            qa_band = qa_img[0]
            
            # Decode QA band
            valid_mask = decode_qa_pixel(qa_band)
            
            # Apply mask to data (set invalid pixels to NaN)
            data = out_img[0].astype("float32")
            data[~valid_mask] = np.nan
            
            return data, out_meta, valid_mask
    else:
        # No QA band available - return data as-is
        return out_img[0].astype("float32"), out_meta, None

# EXTRACT DATE FROM FILENAME
def extract_date_from_filename(filename):
    parts = filename.split('_')
    for part in parts:
        if len(part) == 8 and part.isdigit():
            year = int(part[:4])
            month = int(part[4:6])
            day = int(part[6:8])
            return datetime(year, month, day)
    return None

# FIND ALL LANDSAT SCENES FOR A GIVEN YEAR
def find_all_landsat_scenes(year):
    folder = os.path.join(BASE_DIR, f"{year}")
    if not os.path.exists(folder):
        return []
    
    # Find all unique scene identifiers
    red_files = glob.glob(os.path.join(folder, f"*{BANDS['red']}"))
    scenes = []
    
    for red_file in red_files:
        filename = os.path.basename(red_file)
        scene_id = filename.replace(f"_{BANDS['red']}", "")
        date = extract_date_from_filename(filename)
        if date:
            scenes.append({
                'scene_id': scene_id,
                'date': date,
                'year': date.year,
                'month': date.month,
                'folder': folder
            })
    
    return sorted(scenes, key=lambda x: x['date'])

# FIND LANDSAT FILES FOR A SPECIFIC SCENE
def find_scene_files(scene_info):
    scene_files = {}
    missing_bands = []
    
    for key, suffix in BANDS.items():
        file_path = os.path.join(scene_info['folder'], f"{scene_info['scene_id']}_{suffix}")
        if os.path.exists(file_path):
            scene_files[key] = file_path
        elif key != 'qa':  # QA band is optional
            missing_bands.append(key)
    
    return scene_files, missing_bands

# COMPUTE INDICES
def calc_indices(red, nir, swir):
    """Calculate NDVI and NDBI from Landsat bands with proper scaling"""
    # Apply scaling factors for Landsat Collection 2
    red = red * 0.0000275 - 0.2
    nir = nir * 0.0000275 - 0.2
    swir = swir * 0.0000275 - 0.2
    
    # Calculate indices (handle NaN from cloud masking)
    ndvi = np.where(~np.isnan(red) & ~np.isnan(nir),
                    (nir - red) / (nir + red + 1e-10),
                    np.nan)
    ndbi = np.where(~np.isnan(swir) & ~np.isnan(nir),
                    (swir - nir) / (swir + nir + 1e-10),
                    np.nan)
    
    return ndvi, ndbi

# CLASSIFY DEFORESTED AREAS
def classify_deforested(ndbi, threshold=0.0):
    """Classify deforested areas based on NDBI threshold"""
    result = np.zeros_like(ndbi, dtype="uint8")
    result[ndbi > threshold] = 1
    result[np.isnan(ndbi)] = 255  # Mark masked pixels as no-data
    return result

# ============================================================
# MAIN PROCESSING LOOP WITH CLOUD MASKING
# ============================================================

stats = []
processed_scenes = []
cloud_stats = []

print("\n" + "="*70)
print("üõ∞Ô∏è  PROCESSING LANDSAT SCENES WITH CLOUD/SHADOW MASKING")
print("="*70)

# Collect all scenes from all years
all_scenes = []
for year in YEARS:
    scenes = find_all_landsat_scenes(year)
    all_scenes.extend(scenes)

print(f"‚úì Found {len(all_scenes)} total scenes across {len(YEARS)} years")

for scene in tqdm(all_scenes, desc="Processing Landsat Scenes"):
    scene_files, missing_bands = find_scene_files(scene)
    
    if missing_bands:
        print(f"‚ö† Missing bands for {scene['date'].strftime('%Y-%m-%d')}: {missing_bands}, skipping...")
        continue
    
    try:
        with rasterio.open(scene_files['red']) as src:
            raster_crs = src.crs
        
        # Get QA band path
        qa_path = scene_files.get('qa', None)
        
        # Load bands with cloud masking
        band_data = {}
        valid_mask_all = None
        
        for key in ['red', 'nir', 'swir', 'blue', 'green']:
            if key in scene_files:
                arr, meta, valid_mask = clip_to_aoi_with_qa(scene_files[key], qa_path, aoi, raster_crs)
                band_data[key] = arr
                if valid_mask is not None and valid_mask_all is None:
                    valid_mask_all = valid_mask

        # Calculate cloud coverage statistics
        if valid_mask_all is not None:
            total_pixels = valid_mask_all.size
            valid_pixels = np.sum(valid_mask_all)
            cloud_coverage = 100 * (1 - valid_pixels / total_pixels)
        else:
            cloud_coverage = 0.0
            valid_pixels = np.sum(~np.isnan(band_data['red']))

        # Skip scenes with >50% cloud cover
        if cloud_coverage > 50:
            print(f"‚õÖ Skipping {scene['date'].strftime('%Y-%m-%d')}: {cloud_coverage:.1f}% cloud cover")
            cloud_stats.append({
                'date': scene['date'],
                'cloud_coverage': cloud_coverage,
                'processed': False
            })
            continue

        red, nir, swir = band_data["red"], band_data["nir"], band_data["swir"]
        ndvi, ndbi = calc_indices(red, nir, swir)
        deforested = classify_deforested(ndbi)

        # Create filename with date
        date_str = scene['date'].strftime('%Y%m%d')
        
        # Save outputs with proper nodata handling
        meta.update(dtype=rasterio.float32, count=1, nodata=np.nan)
        with rasterio.open(os.path.join(OUTPUT_DIR, f"NDVI_{date_str}.tif"), "w", **meta) as dst:
            dst.write(ndvi, 1)
        with rasterio.open(os.path.join(OUTPUT_DIR, f"NDBI_{date_str}.tif"), "w", **meta) as dst:
            dst.write(ndbi, 1)

        meta.update(dtype=rasterio.uint8, nodata=255)
        with rasterio.open(os.path.join(OUTPUT_DIR, f"Deforested_{date_str}.tif"), "w", **meta) as dst:
            dst.write(deforested, 1)

        # Calculate statistics (excluding masked pixels)
        deforested_pixels = np.sum(deforested == 1)
        deforested_area_km2 = deforested_pixels * 900 / 1e6  # 30m x 30m = 900 m¬≤
        
        stats.append({
            'Date': scene['date'], 
            'Year': scene['year'], 
            'Month': scene['month'], 
            'Deforested_Pixels': deforested_pixels, 
            'Deforested_Area_km2': deforested_area_km2,
            'Valid_Pixels': int(valid_pixels),
            'Cloud_Coverage': cloud_coverage
        })
        processed_scenes.append(scene)
        cloud_stats.append({
            'date': scene['date'],
            'cloud_coverage': cloud_coverage,
            'processed': True
        })
        
    except Exception as e:
        print(f"‚ùå Error processing {scene['date'].strftime('%Y-%m-%d')}: {str(e)}")
        continue

print(f"\n‚úÖ Processing complete! Successfully processed {len(stats)} scenes")
if stats:
    print(f"   Average cloud coverage: {np.mean([s['Cloud_Coverage'] for s in stats]):.1f}%")

# ============================================================
# SAVE STATISTICS & VISUALIZATIONS
# ============================================================

if not stats:
    print("‚ö† No data was successfully processed!")
else:
    # SAVE STATISTICS WITH CLOUD COVERAGE INFO
    df = pd.DataFrame(stats)
    csv_path = os.path.join(OUTPUT_DIR, "Monthly_Deforestation_Stats.csv")
    df.to_csv(csv_path, index=False)
    print(f"\n‚úì Statistics saved to {csv_path}")
    print(f"\nFirst 10 records:")
    print(df.head(10))
    
    # Save cloud coverage statistics
    cloud_df = pd.DataFrame(cloud_stats)
    cloud_csv = os.path.join(OUTPUT_DIR, "Cloud_Coverage_Stats.csv")
    cloud_df.to_csv(cloud_csv, index=False)
    print(f"‚úì Cloud coverage statistics saved to {cloud_csv}")

    # PLOT MONTHLY DEFORESTATION TREND WITH CLOUD INFO
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 10))
    
    # Plot 1: Deforestation trend
    ax1.plot(df["Date"], df["Deforested_Area_km2"], marker='o', color='crimson', 
             linewidth=2, markersize=5, alpha=0.7, label='Deforested Area')
    ax1.set_title("Monthly Deforested Area - Matale District (Cloud-Masked)", 
                  fontsize=14, fontweight='bold')
    ax1.set_xlabel("Date", fontsize=12)
    ax1.set_ylabel("Deforested Area (km¬≤)", fontsize=12)
    ax1.grid(True, alpha=0.3)
    ax1.legend()
    ax1.tick_params(axis='x', rotation=45)
    
    # Plot 2: Cloud coverage
    ax2.bar(df["Date"], df["Cloud_Coverage"], color='lightblue', alpha=0.6, label='Cloud Coverage')
    ax2.axhline(y=50, color='red', linestyle='--', linewidth=2, alpha=0.5, label='50% Threshold')
    ax2.set_title("Cloud Coverage per Scene", fontsize=14, fontweight='bold')
    ax2.set_xlabel("Date", fontsize=12)
    ax2.set_ylabel("Cloud Coverage (%)", fontsize=12)
    ax2.grid(True, alpha=0.3)
    ax2.legend()
    ax2.tick_params(axis='x', rotation=45)
    
    plt.tight_layout()
    plt.savefig(os.path.join(OUTPUT_DIR, "Monthly_Deforestation_with_Clouds.png"), dpi=200, bbox_inches='tight')
    plt.show()

    # PLOT YEARLY AGGREGATED TREND
    yearly_stats = df.groupby('Year')['Deforested_Area_km2'].agg(['mean', 'min', 'max', 'std']).reset_index()
    
    plt.figure(figsize=(12,6))
    plt.errorbar(yearly_stats['Year'], yearly_stats['mean'], 
                yerr=yearly_stats['std'], marker='o', color='darkred', 
                linewidth=2, markersize=6, capsize=5)
    plt.fill_between(yearly_stats['Year'], yearly_stats['min'], yearly_stats['max'], 
                    alpha=0.2, color='red', label='Min-Max Range')
    plt.title("Yearly Deforested Area Statistics (Mean ¬± Std)", fontsize=14, fontweight='bold')
    plt.xlabel("Year", fontsize=12)
    plt.ylabel("Deforested Area (km¬≤)", fontsize=12)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig(os.path.join(OUTPUT_DIR, "Yearly_Deforestation_Stats.png"), dpi=200, bbox_inches='tight')
    plt.show()

    # SEASONAL ANALYSIS
    df['Season'] = df['Month'].map({
        12: 'Winter', 1: 'Winter', 2: 'Winter',
        3: 'Spring', 4: 'Spring', 5: 'Spring',
        6: 'Summer', 7: 'Summer', 8: 'Summer',
        9: 'Fall', 10: 'Fall', 11: 'Fall'
    })
    
    seasonal_stats = df.groupby('Season')['Deforested_Area_km2'].agg(['mean', 'std']).reset_index()
    
    plt.figure(figsize=(10,6))
    plt.bar(seasonal_stats['Season'], seasonal_stats['mean'], 
           yerr=seasonal_stats['std'], capsize=5, color=['lightblue', 'lightgreen', 'orange', 'lightcoral'])
    plt.title("Seasonal Deforestation Analysis (Cloud-Masked)", fontsize=14, fontweight='bold')
    plt.xlabel("Season", fontsize=12)
    plt.ylabel("Mean Deforested Area (km¬≤)", fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig(os.path.join(OUTPUT_DIR, "Seasonal_Deforestation.png"), dpi=200, bbox_inches='tight')
    plt.show()

    # CREATE CHANGE MAP (First vs Last available scene)
    if len(processed_scenes) >= 2:
        try:
            first_scene = processed_scenes[0]
            last_scene = processed_scenes[-1]
            
            first_date = first_scene['date'].strftime('%Y%m%d')
            last_date = last_scene['date'].strftime('%Y%m%d')
            
            deforested_first = rasterio.open(os.path.join(OUTPUT_DIR, f"Deforested_{first_date}.tif")).read(1)
            deforested_last = rasterio.open(os.path.join(OUTPUT_DIR, f"Deforested_{last_date}.tif")).read(1)

            change = deforested_last.astype('int16') - deforested_first.astype('int16')
            # Update metadata for int8 change map with valid nodata value
            meta.update(dtype=rasterio.int8, nodata=-128)
            with rasterio.open(os.path.join(OUTPUT_DIR, f"DeforestationChange_{first_date}_{last_date}.tif"), "w", **meta) as dst:
                dst.write(change.astype("int8"), 1)

            print(f"\n‚úì Deforestation change map created: DeforestationChange_{first_date}_{last_date}.tif")

            plt.figure(figsize=(10,8))
            plt.imshow(change, cmap='RdBu_r', vmin=-1, vmax=1)
            plt.title(f"Deforestation Change: {first_scene['date'].strftime('%Y-%m-%d')} to {last_scene['date'].strftime('%Y-%m-%d')}\nRed = New Forest Loss, Blue = Recovery", 
                     fontsize=14, fontweight='bold')
            plt.colorbar(label='Change', shrink=0.8)
            plt.axis('off')
            plt.tight_layout()
            plt.savefig(os.path.join(OUTPUT_DIR, "Deforestation_Change_Map.png"), dpi=200, bbox_inches='tight')
            plt.show()
            
        except Exception as e:
            print(f"‚ö† Could not create change map: {str(e)}")
    else:
        print("‚ö† Cannot create change map - insufficient data")

print("\n" + "="*70)
print(f"‚úÖ Analysis Complete! Processed {len(stats)} scenes")
print("="*70)
```

## Temporal Compositing - Cloud Gap Filling

This section creates temporal composite products from all monthly scenes to reduce cloud gaps and improve data quality:

### üéØ **Compositing Strategy**
- **Median Compositing**: Robust to outliers and cloud contamination
- **Observation Count**: Track data quality/temporal density
- **Majority Voting**: Consensus-based deforestation mapping
- **Gap Filling**: Intelligent interpolation for small cloud gaps
- **Optional SAR Fusion**: Fill persistent gaps using Sentinel-1 radar data

### üìä **Output Products**
1. `NDVI_median.tif` - Temporal median NDVI (cloud-free composite)
2. `NDVI_obs_count.tif` - Number of valid observations per pixel
3. `Deforested_majority.tif` - Majority vote deforestation map (speckle-cleaned)
4. `NDVI_median_filled.tif` - Gap-filled NDVI using neighbor interpolation
5. `NDVI_median_sarfilled.tif` - SAR-assisted gap filling (if Sentinel-1 available)

```python
import os
import glob
import numpy as np
import rasterio
from rasterio.enums import Resampling
from tqdm import tqdm
from scipy import ndimage

print("\n" + "="*70)
print("üåç TEMPORAL COMPOSITE GENERATION - CLOUD GAP FILLING")
print("="*70)

# CONFIG - update base paths if different
BASE_DIR = r"D:\Satellite Image Processing\Deforestation_Matale"
PROCESSED_DIR = os.path.join(BASE_DIR, "Processed_Monthly")
OUTPUT_DIR = os.path.join(BASE_DIR, "Processed_Monthly", "Composites")
os.makedirs(OUTPUT_DIR, exist_ok=True)

def list_tifs(prefix):
    """Find all TIF files matching prefix pattern"""
    return sorted(glob.glob(os.path.join(PROCESSED_DIR, f"{prefix}_*.tif")))

def read_stack(files):
    """Read multiple TIFs into a 3D array stack"""
    if len(files) == 0:
        raise ValueError("No files found for pattern")
    # Read first to get profile/shape
    with rasterio.open(files[0]) as src0:
        meta = src0.meta.copy()
        shape = (len(files),) + src0.read(1).shape
    stack = np.full(shape, np.nan, dtype=np.float32)
    for i, f in enumerate(tqdm(files, desc="Reading files", leave=False)):
        with rasterio.open(f) as src:
            data = src.read(1).astype(np.float32)
            # convert nodata markers to NaN
            if src.nodata is not None:
                data[data == src.nodata] = np.nan
            # For deforested layers, 255 = nodata
            if 'Deforested' in f:
                data[data == 255] = np.nan
            stack[i] = data
    return stack, meta

def save_raster(array, out_path, meta, nodata=np.nan):
    """Save array as GeoTIFF with proper nodata handling"""
    meta2 = meta.copy()
    meta2.update(count=1, dtype='float32')
    if np.isnan(nodata):
        meta2.update(nodata=None)
    else:
        meta2.update(nodata=nodata)
    with rasterio.open(out_path, 'w', **meta2) as dst:
        dst.write(array.astype(np.float32), 1)

# ============================================================
# 1) NDVI MEDIAN COMPOSITE + OBSERVATION COUNT
# ============================================================
print("\nüìä Processing NDVI temporal composite...")
ndvi_files = list_tifs("NDVI")
print(f"‚úì Found {len(ndvi_files)} NDVI scenes")

if len(ndvi_files) > 0:
    ndvi_stack, ndvi_meta = read_stack(ndvi_files)
    
    # Calculate median (robust to outliers/clouds)
    ndvi_median = np.nanmedian(ndvi_stack, axis=0)
    
    # Count valid observations per pixel
    ndvi_obs_count = np.sum(~np.isnan(ndvi_stack), axis=0).astype(np.uint16)
    
    # Save outputs
    save_raster(ndvi_median, os.path.join(OUTPUT_DIR, "NDVI_median.tif"), ndvi_meta, nodata=np.nan)
    save_raster(ndvi_obs_count, os.path.join(OUTPUT_DIR, "NDVI_obs_count.tif"), ndvi_meta, nodata=0)
    
    # Statistics
    valid_pixels = np.sum(~np.isnan(ndvi_median))
    total_pixels = ndvi_median.size
    coverage = (valid_pixels / total_pixels) * 100
    avg_obs = np.nanmean(ndvi_obs_count[ndvi_obs_count > 0])
    
    print(f"‚úì NDVI median composite created")
    print(f"  - Valid pixel coverage: {coverage:.1f}%")
    print(f"  - Average observations per pixel: {avg_obs:.1f}")
    print(f"  - NDVI range: [{np.nanmin(ndvi_median):.3f}, {np.nanmax(ndvi_median):.3f}]")
else:
    print("‚ö† No NDVI files found - skipping NDVI composite")
    ndvi_median = None
    ndvi_meta = None

# ============================================================
# 2) DEFORESTED MAJORITY VOTE (WITH SPECKLE CLEANING)
# ============================================================
print("\nüå≤ Processing deforestation majority composite...")
def_files = list_tifs("Deforested")
print(f"‚úì Found {len(def_files)} deforestation scenes")

if len(def_files) > 0:
    def_stack, def_meta = read_stack(def_files)
    
    # Convert to binary: 1 = deforested, 0 = forest, NaN = masked
    def_stack_bin = np.where(np.isnan(def_stack), np.nan, np.where(def_stack == 1, 1.0, 0.0))
    
    # Count valid observations
    valid_count = np.sum(~np.isnan(def_stack_bin), axis=0)
    
    # Calculate mean (fraction of scenes voting deforested)
    mean_def = np.nanmean(def_stack_bin, axis=0)
    
    # Majority vote: >50% of scenes vote deforested
    majority = np.where(valid_count == 0, np.nan, (mean_def > 0.5).astype(np.float32))
    
    # Morphological filtering to remove speckle noise
    print("‚úì Applying morphological cleaning (remove salt-and-pepper noise)...")
    struct = np.ones((3,3))
    maj_bin = np.where(np.isnan(majority), 0, majority).astype(bool)
    
    # Binary opening (erosion then dilation) - removes small isolated deforested pixels
    maj_clean = ndimage.binary_opening(maj_bin, structure=struct)
    
    # Binary closing (dilation then erosion) - fills small holes in deforested areas
    maj_clean = ndimage.binary_closing(maj_clean, structure=struct)
    
    # Restore NaN mask
    majority_clean = np.where(valid_count == 0, np.nan, maj_clean.astype(np.float32))
    
    # Save output
    save_raster(majority_clean, os.path.join(OUTPUT_DIR, "Deforested_majority.tif"), def_meta, nodata=np.nan)
    
    # Statistics
    deforested_pixels = np.sum(majority_clean == 1)
    deforested_area_km2 = deforested_pixels * 900 / 1e6  # 30m x 30m = 900 m¬≤
    valid_pixels = np.sum(~np.isnan(majority_clean))
    deforested_pct = (deforested_pixels / valid_pixels) * 100 if valid_pixels > 0 else 0
    
    print(f"‚úì Majority vote deforestation map created")
    print(f"  - Deforested area: {deforested_area_km2:.2f} km¬≤")
    print(f"  - Deforested percentage: {deforested_pct:.1f}% of valid area")
    print(f"  - Average vote confidence: {np.nanmean(mean_def[mean_def > 0]):.2f}")
else:
    print("‚ö† No Deforested_*.tif found - skipping majority vote")

# ============================================================
# 3) GAP FILLING - NEIGHBOR INTERPOLATION FOR SMALL HOLES
# ============================================================
if ndvi_median is not None:
    print("\nüîß Applying intelligent gap filling for small cloud gaps...")
    
    def fill_small_gaps(arr, max_hole_size_pixels=100):
        """Fill small NaN gaps using mean of neighboring valid pixels"""
        mask_nan = np.isnan(arr)
        labeled, n = ndimage.label(mask_nan)
        out = arr.copy()
        filled_count = 0
        
        for lab in range(1, n+1):
            coords = np.argwhere(labeled == lab)
            hole_size = coords.shape[0]
            
            # Only fill small holes
            if hole_size <= max_hole_size_pixels:
                # Dilate to get neighboring pixels
                dil = ndimage.binary_dilation(labeled == lab, iterations=3)
                neighbors = out[dil & ~np.isnan(out)]
                
                if neighbors.size > 0:
                    # Fill with neighbor mean
                    out[labeled == lab] = np.nanmean(neighbors)
                    filled_count += 1
        
        return out, filled_count, n
    
    ndvi_filled, filled_holes, total_holes = fill_small_gaps(ndvi_median, max_hole_size_pixels=200)
    save_raster(ndvi_filled, os.path.join(OUTPUT_DIR, "NDVI_median_filled.tif"), ndvi_meta, nodata=np.nan)
    
    # Calculate improvement
    original_coverage = (np.sum(~np.isnan(ndvi_median)) / ndvi_median.size) * 100
    filled_coverage = (np.sum(~np.isnan(ndvi_filled)) / ndvi_filled.size) * 100
    improvement = filled_coverage - original_coverage
    
    print(f"‚úì Gap filling completed")
    print(f"  - Total cloud gaps found: {total_holes}")
    print(f"  - Small gaps filled: {filled_holes}")
    print(f"  - Coverage improvement: {improvement:.2f}% ({original_coverage:.1f}% ‚Üí {filled_coverage:.1f}%)")

# ============================================================
# 4) OPTIONAL: SAR FUSION FOR PERSISTENT CLOUD GAPS
# ============================================================
print("\nüõ∞Ô∏è Checking for Sentinel-1 SAR data...")
s1_vh_files = list_tifs("S1_VH")

if len(s1_vh_files) > 0:
    print(f"‚úì Found {len(s1_vh_files)} Sentinel-1 VH scenes")
    print("‚úì Applying SAR-assisted gap filling...")
    
    s1_stack, s1_meta = read_stack(s1_vh_files)
    s1_vh_median = np.nanmedian(s1_stack, axis=0)
    
    # Simple gap-fill heuristic: where NDVI_median is NaN and S1 indicates low backscatter (dense forest),
    # fill NDVI with a high value (tune threshold)
    vh_thresh = np.nanpercentile(s1_vh_median, 40)  # tune as needed
    fill_mask = np.isnan(ndvi_median) & (s1_vh_median < vh_thresh)
    ndvi_sar_filled = ndvi_median.copy()
    ndvi_sar_filled[fill_mask] = 0.7  # assign high-NDVI proxy for forest (tune)
    
    save_raster(ndvi_sar_filled, os.path.join(OUTPUT_DIR, "NDVI_median_sarfilled.tif"), ndvi_meta, nodata=np.nan)
    
    sar_filled_pixels = np.sum(fill_mask)
    sar_filled_area_km2 = sar_filled_pixels * 900 / 1e6
    
    print(f"‚úì SAR fusion completed")
    print(f"  - Pixels filled using SAR: {sar_filled_pixels:,}")
    print(f"  - Area filled: {sar_filled_area_km2:.2f} km¬≤")
else:
    print("‚ö† No Sentinel-1 VH files found (expected format: S1_VH_*.tif)")
    print("  ‚Üí Skipping SAR fusion - optical-only composites created")

# ============================================================
# SUMMARY & VISUALIZATION
# ============================================================
print("\n" + "="*70)
print("‚úÖ TEMPORAL COMPOSITING COMPLETE!")
print("="*70)
print(f"üìÅ Output directory: {OUTPUT_DIR}")
print("\nüì¶ Generated Products:")
print("  ‚Ä¢ NDVI_median.tif - Temporal median NDVI (robust composite)")
print("  ‚Ä¢ NDVI_obs_count.tif - Observation count per pixel")
print("  ‚Ä¢ Deforested_majority.tif - Consensus deforestation map")
print("  ‚Ä¢ NDVI_median_filled.tif - Gap-filled NDVI (neighbor interpolation)")
if len(s1_vh_files) > 0:
    print("  ‚Ä¢ NDVI_median_sarfilled.tif - SAR-assisted gap filling")

print("\nüéØ Key Benefits:")
print("  ‚úì Reduced cloud contamination through temporal aggregation")
print("  ‚úì Improved spatial coverage via intelligent gap filling")
print("  ‚úì Robust deforestation consensus from majority voting")
print("  ‚úì Speckle noise reduction through morphological filtering")
print("  ‚úì Quantified data quality (observation counts)")

print("\nüí° Next Steps:")
print("  ‚Üí Use these composites for improved deforestation risk analysis")
print("  ‚Üí Feed into machine learning models for better predictions")
print("  ‚Üí Create multi-temporal change detection using gap-free data")
print("="*70)
```
